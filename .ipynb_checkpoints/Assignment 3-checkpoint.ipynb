### 1. Look at the big picture 
We want to build a machine learning model from this data that can predict the type of heart defect for a new set of measurements. But before wqe can apply our model to new measurements, we need to know whether it actually works - that is, whether we should trust its predictions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sn
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.feature_selection import RFE, VarianceThreshold, chi2, SelectKBest
```

### 2. Get the data

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
missing_values = ["?", "n/a", "--"]
heartData_uncleaned = pd.read_csv(url, na_values=missing_values, header=None)
heartData_uncleaned = pd.DataFrame(heartData_uncleaned)
heartData_uncleaned = heartData_uncleaned.set_axis(["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num"], axis='columns', inplace=False)
```

### 4. Prepare the data for Machine Learning Algorithms

```python
heartData_uncleaned.isnull().values.any()
#Get the total number of nulls
heartData_uncleaned.isnull().sum().sum()
heartData_uncleaned.count()
```

```python
heartData = heartData_uncleaned.dropna()
heartData.count()
```

Unfortunately, we cannot use the data we used to build the model to evalute it. This is because our model can always simply remember the whole training set, and will therefore always predict the correct label for any point in the training set. To assess the models performance, we show it new data (data that it hasnt seen before) for which we have labels. This is usually done by splitting the labeled data we have collected (here, our 300 heart data) into three parts -->
    
    1. Training
    2. Validation
    3. Testing
    
Validation is usually required for Neural Networks and tuning their hyper parameters. Lets split the data-set in two.

```python
#Splitting the data in 2
train, test = train_test_split(heartData, test_size=0.3)
#computer the mean value per feature on the training set
mean_on_train = train.mean(axis=0)
mean_on_test = test.mean(axis=0)
heartData_uncleaned.head()
c.get_dtype_counts()
```

### 5. Select a model and train it.

Logistic Regression is used for prediction of outcome. In logistic regression the dependent variable is always binary.

#### Multi-Class Logistic Regression
Lets start with Logistic Regression. We will try to classify which of the 5 heart problems the patient has.

```python
X_train = train.iloc[:,0:13]
y_train = train.iloc[:,-1]
X_test = test.iloc[:,0:13]
y_test = test.iloc[:,-1]
```

```python
logreg = LogisticRegression().fit(X_train, y_train)
```

```python
print("Training set score: {:3f}".format(logreg.score(X_train, y_train)))
print("Test set score: {:3f}".format(logreg.score(X_test,y_test)))
```

#### Binary-Class Logistic Regression
Lets change the problem to a binary class classifier

```python
train_binary = train.copy()
test_binary = test.copy()
```

```python
for index, row in train_binary.iterrows():
    ifor_val = 1;
    if row['num'] != 0:
        train_binary.set_value(index, 'num', ifor_val)
        
for index, row in test_binary.iterrows():
    ifor_val = 1;
    if row['num'] != 0:
        test_binary.set_value(index, 'num', ifor_val)

X_train_binary = train_binary.iloc[:,0:13]
y_train_binary = train_binary.iloc[:,-1]
X_test_binary = test_binary.iloc[:,0:13]
y_test_binary = test_binary.iloc[:,-1]
```

```python
logregBinary = LogisticRegression().fit(X_train_binary, y_train_binary)

print("Training set score: {:3f}".format(logregBinary.score(X_train_binary, y_train_binary)))
print("Test set score: {:3f}".format(logregBinary.score(X_test_binary, y_test_binary)))
y_pred = logregBinary.predict(X_test_binary)
print(sklearn.metrics.accuracy_score(y_test_binary, y_pred))
```

### 6. Fine Tune your Model

     Feature Selection is a process where you automatically select those features in your data that contribute most to the prediction variable or output in which you are interested. Having irrelevant features in your data can decrease the accuracy of many models, especially linear algorithms like linear and logistic regression.

```python
cm = confusion_matrix(y_test_binary, y_pred)
conf_matrix = pd.DataFrame(data=cm, columns = ['Predicted:0','Predicted:1'], index=['Actual:0','Actual:1'])
plt.figure(figsize=(8,5))
sn.heatmap(conf_matrix, annot=True, fmt='d',cmap="YlGnBu")
```

The confusion matrix shows 76 correct predictions and 14 incorrect ones.

Trying out different values for the regularizer:

```python
logregBinary100 = LogisticRegression(C=100).fit(X_train_binary, y_train_binary)

print("Training set score: {:3f}".format(logregBinary100.score(X_train_binary, y_train_binary)))
print("Test set score: {:3f}".format(logregBinary100.score(X_test_binary, y_test_binary)))
y_pred100 = logregBinary100.predict(X_test_binary)
print(sklearn.metrics.accuracy_score(y_test_binary, y_pred100))
```

The Test set score does not change. A C value of 1 or 100 is sufficient. C values of 0.01 begin to reduce the test set accuracy.

## Neural Network

We will use Multilayer perceptrons for classification and regression.

```python
mlp = MLPClassifier(random_state=42, hidden_layer_sizes=(10,8,4,1), max_iter=1000, alpha=0.9)
mlp.fit(X_train_binary, y_train_binary)
```

```python
print("Accuracy on training set: {:.2f}".format(mlp.score(X_train_binary, y_train_binary)))
print("Accuracy on test set: {:.2f}".format(mlp.score(X_test_binary, y_test_binary)))
```

Lets normalize the data and try again

```python
mean_on_train = X_train_binary.mean(axis=0)
std_on_train = X_train.std(axis=0)
```

```python
X_train_binary_scaled = (X_train_binary - mean_on_train)/std_on_train
X_test_binary_scaled = (X_test_binary - mean_on_train)/std_on_train
```

```python
mlp2 = MLPClassifier(random_state=0, max_iter=1000, hidden_layer_sizes=(11,8,4), alpha=0.01)
mlp2.fit(X_train_binary_scaled, y_train_binary)
```

```python
print("Accuracy on training set: {:.2f}".format(mlp2.score(X_train_binary_scaled, y_train_binary)))
print("Accuracy on test set: {:.2f}".format(mlp2.score(X_test_binary_scaled, y_test_binary)))
y_pred_norm = mlp2.predict(X_test_binary_scaled)
```

With the sklearn MLP function we are achieving 0.86% accuracy. The highest we have achieved so far. We have also learnt the importance of data normalization.

```python
mean_absolute_error(y_test_binary, y_pred_norm)
```

The MAE is even less than the logistic regression model

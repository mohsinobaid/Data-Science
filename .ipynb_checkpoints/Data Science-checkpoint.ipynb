## Main Steps of a DS project:
1. Look at the big picture

Before building a machine learning model it is often a good idea to inspect the data, to see if the task is easily solvable without machine learning, or if the desired information might not be contained in the data.

Additionally, inspecting your data is a good way to find abnormalities and peculiarities. Maybe some of your irises were measured using inches and not centimeters, for example. In the real world, inconsistencies in the data and unexpected measurements are very common.

One of the best ways to inspect data is to visualize it. On way to do this is by using a scatter plot. A scatter plot of the data puts one feature along the x-axis and another along the y-axis, and draws a dot for each data point. Unfortunately, computer screens have only two dimensions, which allows us to plot only two (or maybe 3) features at a time. It is difficult to plot datasets with more than three features this way. One way around this problem is to do a pair plot, which looks at all possible pairs of features. If you have a small set of features (e.g. 4) then it is quite reasonable.

In order to create pairplots, we first convert the numpy array into a pandas DataFrame. pandas has a function to create pair plots called scatter_matrix. The diagonal of this matrix is filled with histograms of each feature.

```python
x = np.linspace(0,2,100)
plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')
plt.xlabel('this is x label')
plt.ylabel('this is y label')
plt.title("A simple plot")
plt.legend()
plt.show()
```

Subsequent calls to plt.plot re-use the current axes and each add another line. Setting the title, legend, and axis also automatically use the current axes and set the title, create the legend, and label the axis respectively.
Typically one finds oneself making the same plots over and over again, but with different data sets, which leads to needing to write specialized functions to do the plotting. The recommended function signature is something like:

```python
def my_plotter(ax, data1, data2, param_dict):
    """"
    A helper function to make a graph
    ax - the axes to draw to
    data1 - xdata
    data2 - ydata
    param_dict - dictionary of kwargs to pass to ax.plot
    """    
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4,100)
fig, ax = plt.subplots(1,1)
#my_plotter(ax,data1,data2,{'marker':'x', 'xlabel':'wee'})
my_plotter(ax,data1,data2,{'marker':'x'})

fig2, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1,data1,data2,{'marker':'x'})
my_plotter(ax2,data3,data4,{'marker':'o'})

fig3, ((ax11, ax12), (ax21, ax22)) = plt.subplots(2,2)
my_plotter(ax11,data1,data2,{'marker':'x'})
my_plotter(ax12,data3,data4,{'marker':'o'})
my_plotter(ax21,data1,data3,{'marker':'x'})
my_plotter(ax22,data1,data4,{'marker':'o'})
```
